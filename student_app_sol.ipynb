## Student Profile Management

### Subtask:

**Reasoning**:
The first instruction is to initialize an empty dictionary for student profiles and implement `add_student_profile` and `view_student_profile` functions. I will start by defining the dictionary and the `add_student_profile` function.



```python
student_profiles = {}

def add_student_profile(name, department, semester):
    if name in student_profiles:
        print(f"Student '{name}' already exists.")
    else:
        student_profiles[name] = {
            'name': name,
            'department': department,
            'semester': semester
        }
        print(f"Student '{name}' added successfully.")

def view_student_profile(name):
    if name in student_profiles:
        profile = student_profiles[name]
        print(f"\nStudent Profile for {profile['name']}:")
        print(f"  Department: {profile['department']}")
        print(f"  Semester: {profile['semester']}")
    else:
        print(f"Student '{name}' not found.")


# Test the functions
add_student_profile("Alice", "Computer Science", "Fall 2023")
add_student_profile("Bob", "Mathematics", "Spring 2024")
view_student_profile("Alice")
view_student_profile("Charlie")
```

## Add and View Assignments

### Subtask:
Implement functions to add assignments for courses and to view assignments for a specific course or student. Assignments should include details like name, due date, and status (e.g., 'pending', 'completed').


**Reasoning**:
I need to initialize the `assignments` structure for existing courses and define the `add_assignment` function as per the instructions.



```python
for course_id in courses:
    if 'assignments' not in courses[course_id]:
        courses[course_id]['assignments'] = {}

def add_assignment(course_id, assignment_name, due_date):
    if course_id not in courses:
        print(f"Course '{course_id}' not found.")
        return

    if assignment_name in courses[course_id]['assignments']:
        print(f"Assignment '{assignment_name}' already exists in course '{courses[course_id]['name']}'.")
    else:
        courses[course_id]['assignments'][assignment_name] = {
            'due_date': due_date,
            'status': 'pending'
        }
        print(f"Assignment '{assignment_name}' added to course '{courses[course_id]['name']}' (Due: {due_date}).")

print("Assignment management functions initialized.")
```

**Reasoning**:
Implement the `view_assignments_for_course` function to display all assignments for a given course, including their due dates and statuses. Then, test both `add_assignment` and `view_assignments_for_course` with various scenarios.



```python
def view_assignments_for_course(course_id):
    if course_id not in courses:
        print(f"Course '{course_id}' not found.")
        return

    course = courses[course_id]
    if course['assignments']:
        print(f"\nAssignments for {course['name']} (ID: {course_id}):")
        for assignment_name, details in course['assignments'].items():
            print(f"  - {assignment_name} (Due: {details['due_date']}, Status: {details['status']})")
    else:
        print(f"No assignments found for {course['name']} (ID: {course_id}).")

# Test the functions
add_assignment("CS101", "Programming Project", "2023-12-15")
add_assignment("MA201", "Problem Set 1", "2023-11-01")
add_assignment("CS101", "Midterm Exam", "2023-11-20")
add_assignment("CS101", "Programming Project", "2023-12-15") # Test duplicate assignment
add_assignment("PH101", "Lab Report", "2023-12-01") # Test non-existent course

view_assignments_for_course("CS101")
view_assignments_for_course("MA201")
view_assignments_for_course("PH101") # Test non-existent course
```

**Reasoning**:
The previous steps covered adding assignments and viewing assignments for a course. To fully address the subtask, I need to implement a function to view assignments for a specific student across all their enrolled courses. This function will iterate through the student's courses and then display assignments for each course, and then test the new function.



```python
def view_assignments_for_student(student_name):
    if student_name not in student_profiles:
        print(f"Student '{student_name}' not found.")
        return

    student = student_profiles[student_name]
    if 'courses' not in student or not student['courses']:
        print(f"Student '{student_name}' is not enrolled in any courses.")
        return

    print(f"\nAssignments for {student_name}:")
    found_assignments = False
    for course_id in student['courses']:
        if course_id in courses and courses[course_id]['assignments']:
            found_assignments = True
            course = courses[course_id]
            print(f"  Course: {course['name']} (ID: {course_id})")
            for assignment_name, details in course['assignments'].items():
                print(f"    - {assignment_name} (Due: {details['due_date']}, Status: {details['status']})")

    if not found_assignments:
        print(f"No assignments found for {student_name} in any enrolled courses.")


# Test the new function
view_assignments_for_student("Alice")
view_assignments_for_student("Bob")
view_assignments_for_student("Charlie") # Test non-existent student
```

```python
general_tasks = {}

def add_general_task(task_id, description, due_date):
    if task_id in general_tasks:
        print(f"Task with ID '{task_id}' already exists.")
    else:
        general_tasks[task_id] = {
            'description': description,
            'due_date': due_date,
            'status': 'pending'
        }
        print(f"Task '{description}' (ID: {task_id}) added successfully.")

def mark_general_task_complete(task_id):
    if task_id not in general_tasks:
        print(f"Task with ID '{task_id}' not found.")
    elif general_tasks[task_id]['status'] == 'completed':
        print(f"Task '{task_id}' is already marked as complete.")
    else:
        general_tasks[task_id]['status'] = 'completed'
        print(f"Task '{general_tasks[task_id]['description']}' (ID: {task_id}) marked as complete.")


print("General task management system initialized.")

# Test the functions
add_general_task("GT001", "Submit Project Proposal", "2023-11-10")
add_general_task("GT002", "Study for Midterm", "2023-11-15")
add_general_task("GT003", "Buy Groceries", "2023-11-05")
add_general_task("GT001", "Duplicate Task Test", "2023-11-11") # Test duplicate task_id

mark_general_task_complete("GT001")
mark_general_task_complete("GT003")
mark_general_task_complete("GT001") # Test marking already complete task
mark_general_task_complete("GT004") # Test non-existent task

print("\nUpdated General Tasks:")
for task_id, task_info in general_tasks.items():
    print(f"  {task_id}: {task_info['description']} (Due: {task_info['due_date']}, Status: {task_info['status']})")
```

**Reasoning**:

```python
from datetime import datetime

def view_upcoming_tasks():
    today = datetime.now().strftime("%Y-%m-%d")
    upcoming_found = False
    print("\nUpcoming General Tasks:")
    for task_id, task_info in general_tasks.items():
        if task_info['status'] == 'pending' and task_info['due_date'] >= today:
            print(f"  {task_id}: {task_info['description']} (Due: {task_info['due_date']}, Status: {task_info['status']})")
            upcoming_found = True
    if not upcoming_found:
        print("  No upcoming tasks.")

def view_overdue_tasks():
    today = datetime.now().strftime("%Y-%m-%d")
    overdue_found = False
    print("\nOverdue General Tasks:")
    for task_id, task_info in general_tasks.items():
        if task_info['status'] == 'pending' and task_info['due_date'] < today:
            print(f"  {task_id}: {task_info['description']} (Due: {task_info['due_date']}, Status: {task_info['status']})")
            overdue_found = True
    if not overdue_found:
        print("  No overdue tasks.")

print("View upcoming and overdue task functions initialized.")
```
